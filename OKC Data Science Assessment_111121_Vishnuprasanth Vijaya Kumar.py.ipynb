```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import math
import random
 
sn.set(color_codes=True)
 
def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)
 
twowins = 0
threewins = 0
twoscores = []
threescores = []
numPoss = 100
 
for i in range(10000):
    twoscore = 0
    threescore = 0
    for j in range(numPoss):
        r1 = random.random()
        r2 = random.random()
        if r1 < .60:
            twoscore += 2.
        if r2 < .40:
            threescore +=3.
        twowins += 1.
    if twoscore < threescore:
        threewins += 1.
    twoscores.append(twoscore)
    threescores.append(threescore)
 
sn.kdeplot(np.array(twoscores))
sn.kdeplot(np.array(threescores))
plt.title("Comparison of 100 possessions of 2's and 3's")
plt.xlabel("Points Scored")
plt.ylabel("Frequency")
plt.show()
```
